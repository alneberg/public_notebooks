```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
bowtie2_df = pd.read_table('kallisto_data/concoct_inputtable.tsv', index_col=0)
```

```python
bowtie2_df.describe()
```

```python
rows = np.random.choice(bowtie2_df.index.values, 10000, replace=False)
sampled_df = bowtie2_df.ix[rows]
```

```python
sampled_df.describe()
```

```python
samples = ["120314", "120322", "120521", "120717", "120820"]
def load_sample_df(sample):
    kallisto_cov_file = "kallisto_data/kallisto_quant/{0}/abundance.txt".format(sample)
    return pd.read_table(kallisto_cov_file, index_col=0)
```

```python
sample_dfs = []
for sample in samples:
    sample_df = load_sample_df(sample)
    sample_dfs.append((sample, sample_df))
kallisto_df = pd.DataFrame(index=sample_df.index)

for sample, sample_df in sample_dfs:
    kallisto_df['kallisto_coverage_{0}'.format(sample)] = 200*sample_df['est_counts'].divide(sample_df['length'])
```

```python
kallisto_df.describe()
```

```python
kallisto_sampled_df = kallisto_df.ix[rows]
```

```python
kallisto_sampled_df.describe()
```

```python
kallisto_sampled_df.to_csv("kallisto_data/kallisto_sampled.csv")
```
