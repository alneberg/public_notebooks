```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
samples = ["120314", "120322", "120521", "120717", "120820"]
```

```python
bowtie2_df = pd.read_table('kallisto_data/bowtie2_sampled.csv', index_col=0, sep=',')
```

```python
kallisto_df = pd.read_table('kallisto_data/kallisto_sampled.csv', index_col=0, sep=',')
```

```python
for sample in samples:
    short_sample = sample[2:]
    kallisto_df['coverage_{0}'.format(sample)] = bowtie2_df['cov_mean_sample_{0}'.format(short_sample)]
kallisto_df['length'] = bowtie2_df['length']
```

```python
sample='120314'
kal_col = 'kallisto_coverage_{0}'.format(sample)
cov_col = 'coverage_{0}'.format(sample)
x = kallisto_df[kal_col]
y = kallisto_df[cov_col]
x_log = np.log(x+1)
y_log = np.log(y+1)
```

```python
mpl.rc("figure", figsize=(20, 16))

plot_df = kallisto_df[[kal_col, cov_col]].copy()
plot_df['kall > 5'] = (x >= 5)
plot_df['cov > 5'] = (y >= 5)
g = sns.FacetGrid(plot_df, col='kall > 5', row='cov > 5', sharex=False, sharey=False)
g.map(plt.scatter, kal_col, cov_col, alpha=0.7)
```

```python
def length_category(x):
    if x < 200:
        return "< 200"
    elif x < 500:
        return "< 500"
    elif x < 1000:
        return "< 1kb"
    elif x < 5000:
        return "< 5kb"
    return "> 5kb"
```

```python
mpl.rc("figure", figsize=(20, 16))

plot_df = kallisto_df[[kal_col, cov_col, 'length']].copy()
plot_df[kal_col] = x_log
plot_df[cov_col] = y_log

plot_df['length_category'] = plot_df['length'].apply(length_category)
plot_df['log_kall > 1'] = (x_log >= 1)
plot_df['log_cov > 1'] = (y_log >= 1)
g = sns.FacetGrid(plot_df, col='log_kall > 1', row='log_cov > 1', sharex=False, sharey=False, hue='length_category')
g.map(plt.scatter, kal_col, cov_col, alpha=0.7)
g.add_legend()
```

```python
plot_df = kallisto_df[kallisto_df['length']>1000][[kal_col, cov_col, 'length']].copy()
plot_df['kall > 5'] = (x >= 5)
plot_df['cov > 5'] = (y >= 5)
plot_df['length_category'] = plot_df['length'].apply(length_category)
g = sns.FacetGrid(plot_df, col='kall > 5', row='cov > 5', sharex=False, sharey=False, hue='length_category')
g.map(plt.scatter, kal_col, cov_col, alpha=0.7)
g.add_legend()
```
