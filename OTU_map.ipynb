#The Actual Classification

## Adaptive threshold

```python
def mag_connection(corr_threshold, corr_difference):
    otu_connected = {}
    for cluster, group_df in df_cluster.groupby('cluster'):
        avg_df_d[cluster] = group_df[sample_cols].multiply(group_df['length'], axis='index').sum().divide(group_df['length'].sum())
        group_df_t = group_df[sample_cols].T
        cluster_thresholds[cluster] = group_df_t.corrwith(avg_df_d[cluster]).quantile(corr_threshold)
        corr = df_t_nonzero.corrwith(avg_df_d[cluster])
        corr_above = corr[corr > cluster_thresholds[cluster]]

        if corr_above.count() == 0:
            otu = None
            # No significant hit
            signal = 0
        elif corr_above.count() == 1:
            otu = corr_above.idxmax()
            signal = 1 # Unique hit
        else:
            corr_above.sort()
            largest = corr_above[-1]
            second_largest = corr_above[-2]
            if largest - second_largest >= corr_difference:
                otu = corr_above.idxmax()
                signal = 2 # Multiple hits, but one distinct
            else:
                otu = None
                signal = 3
        otu_phylum = None
        if otu:
            otu_phylum = df_otu_class["Taxa"].ix[otu]
            if otu_phylum == "":
                otu_phylum = None
        otu_connected[cluster] = {"otu": otu,
                                  "otu_phylum": otu_phylum,
                                  "mag_phylum": df_mag_tax["Taxa"].ix[cluster],
                                  "signal": signal}
    return otu_connected
```

```python
otu_connected = {}
```

```python
corr_thresholds = [0.0, 0.005, 0.01, 0.02, 0.05]
corr_differences = [0, 0.001, 0.01, 0.05]
for corr_threshold, corr_difference in itertools.product(corr_thresholds, corr_differences):
    if (corr_threshold, corr_difference) in otu_connected:
        continue
    otu_connected[(corr_threshold, corr_difference)] = mag_connection(corr_threshold, corr_difference)
```

### Evaluating performance

```python
from sklearn.metrics import precision_score, recall_score
```

## Phylum Level

```python
result_stats = {}
for corr_threshold, corr_difference in itertools.product(corr_thresholds, corr_differences):
    df_result = pd.DataFrame.from_dict(otu_connected[(corr_threshold, corr_difference)], orient='index')
    classified = df_result[df_result["otu_phylum"].notnull()]
    try:
        prec = precision_score(classified["mag_phylum"], classified["otu_phylum"], average='weighted')
        reca = recall_score(classified["mag_phylum"], classified["otu_phylum"], average='weighted')
    except:
        prec = 0
        reca = 0
    recovery = len(classified.index) / float(len(df_result.index))
    result_stats[(corr_threshold, corr_difference)] = {"precision": prec, "recall": reca, "recovery": recovery}
```

```python
CORR_THRESHOLD = 0.005
CORR_DIFFERENCE = 0.001
df_result = pd.DataFrame.from_dict(otu_connected[(CORR_THRESHOLD, CORR_DIFFERENCE)], orient='index')
```

```python
def decide_result(otu_phylum, mag_phylum):
    if otu_phylum == mag_phylum:
        return "TP"
    elif otu_phylum is None:
        return "N"
    elif otu_phylum != mag_phylum:
        return "FP"
    
```

```python
classification_result = {}
for row_id, row in df_result[['otu_phylum', 'mag_phylum']].iterrows():
    classification_result[row_id] = decide_result(row.otu_phylum, row.mag_phylum)
```

```python
df_result['Classification Result'] = pd.Series(classification_result)
```
