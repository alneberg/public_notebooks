```python

fig, ax = plt.subplots()
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for groupby, color in zip(fin_df.groupby('k'), sns.color_palette('hls', n_colors=9)):
    k, group_k = groupby
    ax.plot(np.log(group_k['maxmemory_in_GiB'].astype(float)), 
            np.log(group_k['runtime_sec'].astype(int)),
            marker='o',
            color=color,
            linestyle='',
            ms=6,
            label="k="+str(k),
            alpha=0.7)
ax.legend()
ax.xaxis.set_view_interval(-3, 5)
ax.set_ylabel('Log Runtime (s)')
ax.set_xlabel('Log MaxMemory (GiB)')
plt.show()
```

```python
fin_df['log_maxmemory'] = np.log(fin_df['maxmemory_in_GiB'])
fin_df['log_runtime'] = np.log(fin_df['runtime_sec'])
fig, (ax1, ax2) = plt.subplots(1,2,sharey=False)
ax1.margins(0.05) # Optional, just adds 5% padding to the autoscaling
ax2.margins(0.05)

for (pca, group_pca), color in zip(fin_df.groupby('pca'), sns.color_palette('hls', n_colors=9)):
    ax1.plot(group_pca['k'].astype(int),
         np.log(group_pca['maxmemory_in_GiB'].astype(float)), 
            marker='o',
            linestyle='',
            color=color,
            ms=6,
            label="pca="+str(pca),
            alpha=0.7)
ax1.legend(loc=2)
ax1.set_yscale('log')
ax1.set_title('Log Maxmemory')
ax1.set_ylabel('Maxmemory (GiB)')
ax1.set_xlabel('Kmer length')

for (pca, group_pca), color in zip(fin_df.groupby('pca'), sns.color_palette('hls', n_colors=9)):
    ax2.plot(group_pca['k'].astype(int),
             group_pca['runtime_sec'].astype(int),          
             marker='o',
             color=color,
             linestyle='',
             ms=6,
             label='pca=' + str(pca),
             alpha=0.7)
ax2.legend(loc=2)
ax2.set_yscale('log')
ax2.set_title('Log Runtime')
ax2.set_ylabel('Runtime (s)')
ax2.set_xlabel('Kmer length')

ax.legend()
```

```python
fin_df['log_maxmemory'] = np.log(fin_df['maxmemory_in_GiB'])
fin_df['log_runtime'] = np.log(fin_df['runtime_sec'])
fig, axs = plt.subplots(2,3,sharey=False)

for (cov_matrix, group_matrix), ax_row in zip(fin_df.groupby('cov_matrix_shape'), axs):
    for (cutoff, group_cutoff), ax in zip(group_matrix.groupby('cutoff'), ax_row):
        
        ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
        for (pca, group_pca), color in zip(group_cutoff.groupby('pca'), sns.color_palette('hls', n_colors=9)):
            ax.plot(group_pca['k'].astype(int),
            group_pca['maxmemory_in_GiB'].astype(float), 
                marker='o',
                linestyle='',
                color=color,
                ms=6,
                label="pca="+str(pca),
                alpha=0.5)
        ax.set_yscale('log')
        ax.legend(loc=2)
        ax.set_title(str(cutoff) + ' | ' + str(cov_matrix))
        # Only define ylabel for the leftmost graphs
        if cutoff == 500:
            ax.set_ylabel('Maxmemory (GiB)')
        # Only define xlabel for the lower row
        if cov_matrix == 'full':
            ax.set_xlabel('Kmer length')
        ax.yaxis.set_view_interval(0.09, 700)
fig.suptitle('Maxmemory')
```

```python
above_df = fin_df[fin_df['AdjRand_order'] > 0.1]
fig, axs = plt.subplots(3,1)
ax1, ax2, ax3 = axs
ax1.hist(above_df.pca, bins=[25, 35, 45, 55, 65, 75, 85, 95, 105])
ax1.set_title('pca')

ax2.hist(above_df.k, bins=[0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5,9.5,10.5])
ax2.set_title('Kmer length')

bar_width = 0.35
ax3.bar(0, len(above_df[above_df['cov_matrix_shape'] == 'diag']), bar_width)
ax3.bar(0+bar_width, len(above_df[above_df['cov_matrix_shape'] == 'full']), bar_width)
ax3.set_title('Covariance matrix shape')
xTickMarks = ['diag', 'full']
ax3.set_xticks([bar_width/2, bar_width + bar_width/2])
xtickNames = ax3.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=45, fontsize=10)
fig.suptitle('Runs with AdjRand_order over 0.1 have the following parameters:')
```

This plot was made to see why the adjrand violinplots are timeglass shaped. The value of 'k' seems to have a big influence on the result, but the pca parameter also seems to have some influence.

```python
fig, axs = plt.subplots(2,3,sharey=False)

for (cov_matrix, group_matrix), ax_row in zip(fin_df.groupby('cov_matrix_shape'), axs):
    for (cutoff, group_cutoff), ax in zip(group_matrix.groupby('cutoff'), ax_row):
        
        ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
        for (pca, group_pca), color in zip(group_cutoff.groupby('pca'), sns.color_palette('hls', n_colors=9)):
            ax.plot(group_pca['k'].astype(int),
            group_pca['AdjRand_order'].astype(float), 
                marker='o',
                linestyle='',
                color=color,
                ms=6,
                label="pca="+str(pca),
                alpha=0.5)
        ax.set_yscale('linear')
        ax.legend(loc=1)
        ax.set_title(str(cutoff) + ' | ' + str(cov_matrix))
        ax.xaxis.set_view_interval(1, 13)
        ax.yaxis.set_view_interval(-0.05, 0.4)
        
        # Only define ylabel for the leftmost graphs
        if cutoff == 500:
            ax.set_ylabel('Adjusted Rand Index')
        # Only define xlabel for the lower row
        if cov_matrix == 'full':
            ax.set_xlabel('Kmer length')
fig.suptitle('AdjRand Order')
```
