```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_context("poster")
```

```python
#Four lines corresponding to four genomes
plt.axhline(y=1, xmin=0, xmax=1.0, linewidth=2, color = 'k')
plt.axhline(y=2, xmin=0, xmax=1.0, linewidth=2, color = 'k')
plt.axhline(y=3, xmin=0, xmax=1.0, linewidth=2, color = 'k')
plt.axhline(y=4, xmin=0, xmax=1.0, linewidth=2, color = 'k')
plt.axhline(y=5, xmin=0, xmax=1.0, linewidth=2, color = 'k')

# Plot the cutoff
l = plt.axvline(x=0.85, color='r')
l.set_linestyle('--')

#Zeroeth genome have no good hits
x = [0.6, 0.65,0.74,0.76]
y = [5]*4
plt.scatter(x,y, s=100, c='k')
for i, xy in enumerate(zip(x,y)):
    x,y = xy
    plt.annotate('OTU {0}'.format(i), xy=(x, y), xytext=(x-0.03, y-0.15), fontsize=15, rotation=45)

#First genome is a normal ideal case
x = [0.6, 0.75, 0.95]
y = [4]*3
plt.scatter(x, y, s=100, c='k')
for i, xy in enumerate(zip(x,y)):
    x,y = xy
    plt.annotate('OTU {0}'.format(i+1), xy=(x, y), xytext=(x-0.03, y-0.15), fontsize=15, rotation=45)
    
#Second genome have two good hits, but one that sticks out
x = [0.65, 0.7, 0.88, 0.96]
y = [3]*4
plt.scatter(x, y, s=100, c='k')
for i, xy in enumerate(zip(x,y)):
    x,y = xy
    plt.annotate('OTU {0}'.format(i+1), xy=(x, y), xytext=(x-0.03, y-0.15), fontsize=15, rotation=45)

#Third genome have two good hits and no one sticks out
x = [0.6, 0.7, 0.95, 0.97]
y =  [2]*4
plt.scatter(x, y, s=100, c='k')
for i, xy in enumerate(zip(x,y)):
    x,y = xy
    plt.annotate('OTU {0}'.format(i+1), xy=(x, y), xytext=(x-0.03, y-0.15), fontsize=15, rotation=45)

#Fourth genome have only one good hit, but another one really close
x = [0.53, 0.62, 0.84, 0.86]
y = [1]*4
plt.scatter(x, y, s=100, c='k')
for i, xy in enumerate(zip(x,y)):
    x,y = xy
    plt.annotate('OTU {0}'.format(i+1), xy=(x, y), xytext=(x-0.03, y-0.15), fontsize=15, rotation=45)

plt.xlabel('Pearson Correlation')
plt.yticks( [1,2,3,4,5], ( 'Genome 5','Genome 4', 'Genome 3', 'Genome 2', 'Genome 1') )
plt.ylim(0.4, 5.6)
plt.xlim(0.5, 1.0)

# Annotate thresholds
plt.text(0.82, 0.2, "Threshold", fontsize=17, rotation=45)
t = plt.annotate(
    '', xy=(0.962, 3.15), xycoords='data',
    xytext=(0.902, 3.15), textcoords='data',
    arrowprops={'arrowstyle': '<->', 'linewidth': 2.5, 'color':'red'})

plt.annotate(
    'Difference', xy=(0.902, 3.2), xycoords='data',
    xytext=(5, 0), textcoords='offset points', fontsize=17)
#plt.annotate('y diff', xy=(0.94, 3.1), xytext=(0.93, 3.5), arrowprops=dict(arrowstyle='-[, widthB=4', facecolor='black'))

t = plt.annotate(
    '', xy=(0.972, 2.15), xycoords='data',
    xytext=(0.912, 2.15), textcoords='data',
    arrowprops={'arrowstyle': '<->', 'linewidth': 2.5, 'color':'red'})

plt.annotate(
    'Difference', xy=(0.912, 2.2), xycoords='data',
    xytext=(5, 0), textcoords='offset points', fontsize=17)

ax = plt.axes()
ax2 = ax.twinx()
ax2.grid(False)
ax2.set_ylim(0.4, 5.6)
ax2.set_yticks([1,2,3,4,5])
labels = ax2.set_yticklabels(( 'OTU 4', 'None', 'OTU 4', 'OTU 3', 'None') );
plt.text(s="Assignment", x=0.98, y=5.65, fontdict={'fontsize':18})
plt.savefig("/Users/johannesalneberg/repos/public_notebooks/OTU_data/figures/method_parameters.png")
```
