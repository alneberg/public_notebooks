```python
import pandas as pd
import glob
```

```python
samples = [p.split("/")[2] for p in glob.glob("kallisto_data/kallisto_quant/*/abundance.txt")]
```

```python
def load_sample_df(sample):
    kallisto_cov_file = "kallisto_data/kallisto_quant/{0}/abundance.txt".format(sample)
    return pd.read_table(kallisto_cov_file, index_col=0)
```

```python
sample_dfs = []
for sample in samples:
    sample_df = load_sample_df(sample)
    sample_dfs.append((sample, sample_df))
kallisto_df = pd.DataFrame(index=sample_df.index)

for sample, sample_df in sample_dfs:
    kallisto_df['kallisto_coverage_{0}'.format(sample)] = 200*sample_df['est_counts'].divide(sample_df['length'])
```

```python
kallisto_df.to_csv("kallisto_data/concoct_inputtable_kallisto.tsv", sep="\t", float_format="%.6f")
```
