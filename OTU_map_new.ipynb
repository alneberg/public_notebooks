# The actual classification

```python
def mag_connection(corr_threshold, corr_difference, df_t_nonzero):
    otu_connected = {}
    for cluster, group_df in df_cluster.groupby('cluster'):
        avg_df_d[cluster] = group_df[sample_cols].multiply(group_df['length'], axis='index').sum().divide(group_df['length'].sum())
        group_df_t = group_df[sample_cols].T
        corr = df_t_nonzero.corrwith(avg_df_d[cluster])
        corr_above = corr[corr > corr_threshold]

        corr.sort()
        largest = corr[-1]
        second_largest = corr[-2]

        unit_corr_difference = largest - second_largest
        if unit_corr_difference < corr_difference:
            otu = None
            # Not a significant difference
            if corr_above.count() == 0:
                # Neither a significant difference nor hit
                signal = 0
            elif corr_above.count() == 1:
                # A significant hit but not difference
                signal = -1
            else:
                # Multiple significant hits, but not significant diff
                signal = -2
        else:
            if corr_above.count() == 0:
                otu = None
                # No significant hit, but significant diff
                signal = -3
            elif corr_above.count() == 1:
                otu = corr_above.idxmax()
                signal = 1 # Unique hit
            else:
                otu = corr_above.idxmax()
                signal = 2 # Multiple hits, but one distinct
        otu_phylum = None
        otu_connected[cluster] = {"otu": otu,
                                  "signal": signal,
                                  "2nd_largest": second_largest,
                                  "largest": largest}
        otu_connected[cluster].update(**df_mag_greengenes_tax.ix[cluster])
        if otu:
            for key, value in df_otu_greengenes_tax.ix[otu].iteritems():
                otu_connected[cluster]["otu_" + key] = value
    return otu_connected
```

## Some methods to calculate stats

```python
def precision_score(y_true, y_pred):
    s = y_true == y_pred
    return s[s].count() / float(len(y_true))
```

```python
def fraction_classified(y_true, y_pred):
    return len(y_pred[y_pred.notnull()])/float(len(y_true))
```

```python
def mag_connection_2nd(corr_threshold, corr_difference, df_t_nonzero):
    otu_connected = {}
    for cluster, group_df in df_cluster.groupby('cluster'):
        avg_df_d[cluster] = group_df[sample_cols].multiply(group_df['length'], axis='index').sum().divide(group_df['length'].sum())
        group_df_t = group_df[sample_cols].T
        corr = df_t_nonzero.corrwith(avg_df_d[cluster])
        corr_above = corr[corr > corr_threshold]
        corr_above.sort()
        corr.sort()
        largest = corr[-1]
        second_largest = corr[-2]

        unit_corr_difference = largest - second_largest
        if unit_corr_difference < corr_difference:
            otu = None
            # Not a significant difference
            if corr_above.count() == 0:
                # Neither a significant difference nor hit
                signal = 0
            elif corr_above.count() == 1:
                # A significant hit but not difference
                signal = -1
            else:
                # Multiple significant hits, but not significant diff
                signal = -2
        else:
            if corr_above.count() == 0:
                otu = None
                # No significant hit, but significant diff
                signal = -3
            elif corr_above.count() == 1:
                otu = corr.index[-2]
                signal = 1 # Unique hit
            else:
                otu = corr.index[-2]
                signal = 2 # Multiple hits, but one distinct
        otu_phylum = None
        otu_connected[cluster] = {"otu": otu,
                                  "signal": signal,
                                  "2nd_largest": second_largest,
                                  "largest": largest}
        otu_connected[cluster].update(**df_mag_greengenes_tax.ix[cluster])
        if otu:
            for key, value in df_otu_greengenes_tax.ix[otu].iteritems():
                otu_connected[cluster]["otu_" + key] = value
    return otu_connected
```
