```python
regex = re.compile("[0-9]+")
```

```python
# Parse logs

data = {}
for threshold in ['nm0', 'nm1', 'nm2', 'nm3']:
    fn = "{0}/filtering_logs/{1}.log".format(DATA_LOC, threshold)
    reads_with_good_hit = 0
    reads_kept = 0
    with open(fn) as fh:
        for line in fh:
            if not line.startswith("#"):
                reads_int = [int(x) for x in regex.findall(line)]
                reads = dict(zip(["total", "bad_first", "multiple_good"], reads_int))
                reads_with_good_hit += reads["total"] - reads["bad_first"]
                reads_kept += reads["total"] - reads["bad_first"] - reads["multiple_good"]
    data[threshold] = (reads_with_good_hit, reads_kept)

    
x = []
y = []
names = []
for key, value in data.iteritems():
    x.append(value[1])
    y.append(value[0])
    names.append(key)
```

## Data preprocessing

```python
df_otu_class = pd.read_table("{0}/training_data/SMHI16S_150327_min300.tsv".format(DATA_LOC), sep='\t', index_col=0)
df_otu_class_online = pd.read_table("{0}/otus99num.300_subset_silva_online.csv".format(DATA_LOC), sep=';', index_col=2)
otu_tax = df_otu_class.Taxonomy
```

```python
def greengenes_tax_down_to_order(tax_s):
    """Transform a greengenes taxonomy string from phylum to order level."""
    split_s = tax_s.split(";")
    result_d = {}
    abbrevation_to = {'k': 'Kingdom',
                      'p': 'Phylum',
                      'c': 'Class',
                      'o': 'Order',
                      'f': 'family'}
    for t in split_s:
        unit = t.split('_')
        if len(unit) == 1:
            return result_d
        elif unit[0] == 'f' or unit[0] == 'g':
            return result_d
        result_d[abbrevation_to[unit[0]]] = unit[-1]
    return result_d
```

```python
greengenes_tax_down_to_order("k__Bacteria;p__Proteobacteria;c__Alphaproteobacteria;o__Rickettsiales;f__Pelagibacteraceae;")
```

```python
greengenes_tax_down_to_order("")
```

```python
kingdom = {}
phylum = {}
class_ = {}
order = {}
for otu, greengenes_tax in df_otu_class_online['lca_tax_greengenes'].iteritems():
    taxonomy = greengenes_tax_down_to_order(greengenes_tax)
    if len(taxonomy) == 0:
        kingdom[otu] = None
        phylum[otu] = None
        class_[otu] = None
        order[otu] = None
    if len(taxonomy) > 0:
        kingdom[otu] = taxonomy['Kingdom']
    if len(taxonomy) > 1:
        phylum[otu] = taxonomy['Phylum']
    if len(taxonomy) > 2:
        class_[otu] = taxonomy['Class']
    if len(taxonomy) > 3:
        order[otu] = taxonomy['Order']

df_taxonomy = pd.DataFrame.from_dict({'kingdom': kingdom,
 'phylum': phylum,
 'class': class_,
 'order': order})
```

```python
df_taxonomy.head()
```

```python
df_taxonomy.to_csv("{0}/SMHI16S_subset_green_genes_taxonomy.tsv".format(DATA_LOC), sep='\t')
```

```python
def silva_tax_to_phylum(tax_s):
    """Transform a silva taxonomy string to only phylum level, or class for proteobacteria"""
    split_s = tax_s.split(";")
    if len(split_s) < 2:
        return ""
    else:
        if split_s[1] == "Proteobacteria":
            return split_s[2]
        else:
            return split_s[1]
```

```python
def tax_to_class(tax_s):
    """Transform a silva taxonomy string to class level"""
    split_s = tax_s.split(";")
    if len(split_s) < 4:
        return None
    else:
         return split_s[3]
```

```python
df_otu_class["Taxa"] = otu_tax.apply(tax_to_phylum)
df_otu_class.to_csv("{0}/training_data/SMHI16S_150327_min300_phylum.tsv".format(DATA_LOC), sep='\t')
```

## Hierarchical Clustering

```python
df = pd.read_table("{0}/filtered_nm0/otu_mapping.tsv".format(DATA_LOC), sep='\t', index_col=0)
df_cluster = pd.read_table("{0}/all_bins_cluster2.csv".format(DATA_LOC), sep=',', index_col=0)
```

```python
sample_cols = filter(lambda x: x != 'length', df.columns)
```

```python
df_t = df[sample_cols].T
```

```python
df_t_joined = df_t.copy()
clusters = []
for cluster, group_df in df_cluster.groupby('cluster'):
    clusters.append(cluster)
    df_t_joined[cluster] = group_df[sample_cols].multiply(group_df['length'], axis='index').sum().divide(group_df['length'].sum()).T
```

```python
otus_with_phylum = set(df_otu_class[df_otu_class["Taxa"] != ''].index)
```

```python
non_zero_otus_before_2nd_filtering = [otu for otu, max_val in filter(lambda x: x[1] >= 10.0, df_t.sum().iteritems())]
non_zero_otus = filter(lambda x: x in otus_with_phylum, non_zero_otus_before_2nd_filtering)
```

```python
len(non_zero_otus)
```

```python
y = df_t[non_zero_otus].corr(method='spearman')
```

```python
tree = hierarchy.linkage(y)
```

```python
mpl.rc("figure", figsize=(100, 80))
tree_plot = hierarchy.dendrogram(tree)
```

```python
z = df_t_joined[non_zero_otus + clusters].corr(method='spearman')
tree_joined = hierarchy.linkage(z)
```

```python
mpl.rc("figure", figsize=(100, 80))
tree_plot = hierarchy.dendrogram(tree_joined, labels=z.index.values, leaf_rotation="90")
plt.savefig("test_tree.png")
```
