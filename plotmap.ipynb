```python
#colors = ['red','blue','green','yellow','magenta','purple']
colors = [mpl.colors.rgb2hex(c) for c in sns.color_palette("colorblind")]
def draw_pie(ax,ratios=[0.4,0.3,0.3], X=0, Y=0, size = 1000):
    N = len(ratios)
 
    xy = []
 
    start = 0.
    for ratio in ratios:
        x = [0] + np.cos(np.linspace(2*math.pi*start,2*math.pi*(start+ratio), 30)).tolist()
        y = [0] + np.sin(np.linspace(2*math.pi*start,2*math.pi*(start+ratio), 30)).tolist()
        xy1 = zip(x,y)
        xy.append(xy1)
        start += ratio
 
    for i, xyi in enumerate(xy):
        ax.scatter([X],[Y] , marker=(xyi,0), s=size, facecolor=colors[i] )
```

```python
otu_to_label = {}
actino_clades = {'CL500-29 marine group': 'acIV',
                 'Candidatus Aquiluna': 'Luna',
                 'hgcI clade': 'Luna'
                }
for otu, row in df_otu_greengenes_tax_tmp.ix[all_cluster_otus].iterrows():
    phylum = row['lca_tax_slv'].split(';')[1]
    clade = row['lca_tax_slv'].split(';')[-2]
    if clade == 'uncultured':
        clade = row['lca_tax_slv'].split(';')[-3]
    if phylum == 'Actinobacteria':
        clade = " {0}".format(actino_clades[clade])
    if phylum == 'Verrucomicrobia':
        phylum = 'Verruco.'
        clade = clade.replace("Candidatus", "Cand.")
    otu_to_label[otu] = phylum + '; ' + clade
otu_to_label
```
