```python
import pandas as pd
```

```python
DATA_LOC = "OTU_data"
```

```python
df_otu_class = pd.read_table("{0}/training_data/SMHI16S_150327_min300_phylum.tsv".format(DATA_LOC), sep='\t', index_col=0)
```

Use only the samples named filtrated for abundance filtering.

```python
sample_cols = filter(lambda x: 'Filtrated' in x, df_otu_class.columns)
```

```python
df_otu_class.head()
```

```python
sample_cols
```

```python
non_zero_otus = [otu for otu, sum_val in df_otu_class[sample_cols].T.sum().iteritems() if sum_val > 10.0]
```

```python
len(non_zero_otus)
```

```python
with open("{0}/filtered_otu_list.txt".format(DATA_LOC), 'w') as of:
    for otu in non_zero_otus:
        of.write(otu + '\n')
```
